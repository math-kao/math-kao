#Imports

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

#Pre-processing & Visualization

```python
data=pd.read_csv("cost_revenue_clean.csv")
data.head(3)
```

```python
data.describe()
```

```python
#Features x Target

X=pd.DataFrame(data,columns=['production_budget_usd'])
y=pd.DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
#X=data[['production_budget_usd']]
#X
```

```python
plt.figure(figsize=(8,4),dpi=200)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Film Cost')
plt.ylabel('Global Revenue')

plt.scatter(X,y,alpha=0.5)  #alpha= transparência

plt.ylim(0,3000000000)
plt.xlim(0,450000000)

plt.show()
```

#Training using scikitlearn

```python
regression=LinearRegression()

regression.fit(X,y)
```

```python
regression.coef_  #theta1
```

```python
regression.intercept_ #theta0
```

```python
regression.score(X,y) #R² 54% da variância em y é explicada pelo parâmetro em x
```

```python
#Regression and scatter plot

plt.figure(figsize=(6,2),dpi=200)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Film Cost')
plt.ylabel('Global Revenue')

plt.plot(X,regression.predict(X),linewidth=5,color='red')

plt.scatter(X,y,alpha=0.5)  #alpha= transparência

plt.ylim(0,3000000000)
plt.xlim(0,450000000)

plt.show()
```
